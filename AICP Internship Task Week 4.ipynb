Q.4: Draw boxplot for Calories vs Category to spot outliers and max calories category


```python
import matplotlib.patches as mpatches

plt.figure(figsize=(12, 8))
sns.boxplot(x='Calories', y='Category', data=menu_df)

# Define custom legend patches
legend_patches = [mpatches.Patch(color=color, label=label) for label, color in zip(menu_df['Category'].unique(), sns.color_palette())]

# Add legend
plt.legend(handles=legend_patches, title='Category')

plt.title('Boxplot of Calories by Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.show()
```

Q.6: Draw Stripplot for each
category against the following
attributes ['Calories', 'Total Fat',
'Carbohydrates', 'Dietary
Fiber', 'Sugars', 'Protein',
'Vitamin A (% Daily Value)',
'Vitamin C (% Daily Value)',
'Calcium (% Daily Value)', 'Iron
(% Daily Value)'.


```python
# List of attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 
              'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 
              'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
              'Iron (% Daily Value)']

# Iterate over each attribute
for attribute in attributes:
    plt.figure(figsize=(15, 6))
    
    # Create stripplot for the current attribute
    sns.stripplot(x='Category', y=attribute, data=menu_df, jitter=True)
    
    # Set title and labels
    plt.title(f'Stripplot of {attribute} by Category')
    plt.xlabel('Category')
    plt.ylabel(attribute)
    plt.xticks(rotation=45, ha='right')
    
    plt.show()
```

Q.7: Draw a horizontal bar graph for items in each category against the calories.
Here is one sample for all items in Beef & Pork, you have to do it for items in each category. Also, write
your observation.


```python
plt.figure(figsize=(12, 8))

# Iterate over each category
for category in menu_df['Category'].unique():
    # Filter the DataFrame for items in the current category
    category_df = menu_df[menu_df['Category'] == category]
    
    # Plot a horizontal bar graph for items in the current category against calories
    sns.barplot(x='Calories', y='Item', data=category_df, color='skyblue')
    
    plt.title('Calories of Items in Each Category')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.xlim(0, None)  # Set x-axis limit to start from 0
    
    plt.show()

```

Observations:

By plotting horizontal bar graphs for items in each category against calories, we can easily compare the calorie content of different items within the same category.
The length of the bars represents the calorie content of each item, allowing us to identify which items have higher or lower calorie content relative to others within the same category.
This visualization provides a clear overview of the calorie distribution across different items within each category, facilitating comparison and decision-making for consumers interested in managing their calorie intake or making healthier food choices.
